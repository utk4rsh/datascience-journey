## Predict Flight Delay

```python

# Pandas is a data processing and CSV file I/O library. It creates a data frame for the CSV for processing.
import pandas as pd
from sklearn.preprocessing import LabelEncoder, StandardScaler
import warnings # current version of seaborn generates a bunch of warnings that we'll ignore
warnings.filterwarnings("ignore")
#Plot all matplot graphs inline
%matplotlib inline
# Matplot lib is another Python graph plot library
import matplotlib.pyplot as plt
from sklearn import cross_validation
import numpy as np
from sklearn.linear_model import SGDClassifier
from sklearn.pipeline import Pipeline
from sklearn.grid_search import GridSearchCV
from sklearn.metrics import confusion_matrix
from keras.models import Sequential
from keras.layers.core import Dense, Activation
from keras.utils import np_utils

# Load the train file into a dataframe
train_df = pd.read_csv('data/train_nn.csv', header=0)
```

```python
def hour_ex(x): 
    h = int(str(int(x)).zfill(4)[:2])
    return h
```

### Get the columns which can be treated as features

```python
features_columns_numeric = ['DepTime',\
       'CRSDepTime', 'ArrTime', 'CRSArrTime',\
       'FlightNum', 'ActualElapsedTime', 'CRSElapsedTime',\
       'AirTime', 'ArrDelay', 'Distance', 'TaxiIn', 'TaxiOut']
```

### Feature Scaling

```python
scaler = StandardScaler()
for col in features_columns_numeric:
    scaler.fit(list(train_df[col]))
    train_df[col] = scaler.transform(train_df[col])
train_df.head()
```

### Feature Encoding

```python
features_non_numeric = ['UniqueCarrier','TailNum', 'Origin', 'Dest']
```

```python
le = LabelEncoder()
for col in features_non_numeric:
    le.fit(list(train_df[col]))
    train_df[col] = le.transform(train_df[col])
train_df.head()
```

```python
train_df.drop(['Unnamed: 0', 'Year'])
```

In this section, we add a new column to our data frame, isDelayed, a binary variable:
True, for flights that have > 10 minutes of delay
False, for flights that have <= 10 minutes of delay
We will later use isDelayed as the target/label column in the classification process.

```python
train_df['isDelayed'] = train_df.DepDelay.map(lambda x: x > 10)
train_df.isDelayed.head()
```
