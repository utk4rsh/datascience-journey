```python
import matplotlib.pyplot as plt
%matplotlib inline
import pandas as pd
import csv as csv
from sklearn.preprocessing import LabelEncoder, StandardScaler
import warnings # current version of seaborn generates a bunch of warnings that we'll ignore
warnings.filterwarnings("ignore")
#Plot all matplot graphs inline
%matplotlib inline
# Matplot lib is another Python graph plot library
import matplotlib.pyplot as plt
from sklearn import cross_validation
import numpy as np
from sklearn.linear_model import SGDClassifier, LogisticRegression
from sklearn.pipeline import Pipeline
from sklearn.grid_search import GridSearchCV
from sklearn.metrics import confusion_matrix

# Load the train file into a dataframe
train_df = pd.read_csv('cleaned_train.csv', header=0)
```

### Load the test data as well

```python
# Load the train file into a dataframe
test_df = pd.read_csv('cleaned_test.csv', header=0)
test_df.head()
```

### Get the columns which can be treated as features

```python
columns_numeric = ['Age', 'FamilySize', 'Fare']
```

### Feature Scaling on numeric columns

```python
scaler = StandardScaler()
for col in columns_numeric:
    scaler.fit(list(train_df[col]) + list(test_df[col]))
    train_df[col] = scaler.transform(train_df[col])
    test_df[col] = scaler.transform(test_df[col])
```

```python
train_df.head()
```

```python
test_df.head()
```

### Encoding Non Numeric Columns

```python
features_non_numeric = ['Embarked', 'Sex']
```

```python
le = LabelEncoder()
for col in features_non_numeric:
    le.fit(list(train_df[col]) + list(test_df[col]))
    train_df[col] = le.transform(train_df[col])
    test_df[col] = le.transform(test_df[col])
```

```python
train_df.head()
```

```python
test_df.head()
```

## Writing a classifier

####Drop unwanted features

```python
y_values = list(train_df.Survived)
train_df = train_df.drop(['Unnamed: 0', 'Survived', 'Name', 'PassengerId', 'Ticket', 'SibSp', 'Parch'], axis=1)
train_df.head()
```

```python
ids = test_df['PassengerId'].values
test_df = test_df.drop(['Unnamed: 0',  'Name', 'PassengerId', 'Ticket', 'SibSp', 'Parch'], axis=1)
test_df.head()
```

### Train Test & Validation

```python
X_train, X_test, y_train, y_test = cross_validation.train_test_split(train_df, y_values, test_size=0.25, random_state=0)
print "Training Size ", len(X_train)
print "Test Size ", len(X_test)
```

```python
pipeline = Pipeline([
        ('clf', LogisticRegression()),
    ])

parameters = {
#'clf__alpha': (0.00001, 0.000001),
'clf__penalty': ('l2', 'l1'),
'clf__max_iter': (1000, 2000),
#'clf__class_weight': ('balanced', 'auto', None),
}

grid_search = GridSearchCV(pipeline, parameters, n_jobs=-1)
model = grid_search.fit(X_train, y_train)
```

### Accuracy on train and test

```python
score = model.score(X_train, y_train)
print "Mean accuracy on the train data is: {0}".format(score)
```

```python
score = model.score(X_test, y_test)
print "Mean accuracy on the test data is: {0}".format(score)
```

###Confusion Matrix on test split

```python
from sklearn.metrics import confusion_matrix

def plot_confusion_matrix(cm, title='Confusion matrix', cmap=plt.cm.Blues):
    plt.imshow(cm, interpolation='nearest', cmap=cmap)
    plt.title(title)
    plt.colorbar()
    tick_marks = np.arange(len(np.unique(y_values)))
    plt.xticks(tick_marks, np.unique(y_values), rotation=45)
    plt.yticks(tick_marks, np.unique(y_values))
    plt.tight_layout()
    plt.ylabel('True label')
    plt.xlabel('Predicted label')
 

y_pred = model.predict(X_test)
# Compute confusion matrix
cm = confusion_matrix(y_test, y_pred)
np.set_printoptions(precision=2)
print('Confusion matrix, without normalization')
print(cm)
plt.figure()
plot_confusion_matrix(cm)
```

#### Submission data

```python
print 'Predicting Output for Submission ...'
output = model.predict(test_df).astype(int)

predictions_file = open("submissing.csv", "wb")
open_file_object = csv.writer(predictions_file)
open_file_object.writerow(["PassengerId","Survived"])
open_file_object.writerows(zip(ids, output))
predictions_file.close()
print 'Done.'
```
