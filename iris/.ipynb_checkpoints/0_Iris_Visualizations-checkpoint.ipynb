## Python data visualizations on the Iris dataset

Set of visualizations and data exploration on Iris Dataset. 

```python
# Pandas is a data processing and CSV file I/O library. It creates a data frame for the CSV for processing.
import pandas as pd

# Seaborn is a Python graphing library
import warnings # current version of seaborn generates a bunch of warnings that we'll ignore
warnings.filterwarnings("ignore")
import seaborn as sns
#Plot all matplot graphs inline
%matplotlib inline

# Matplot lib is another Python graph plot library
import matplotlib.pyplot as plt

#Set the style for searborn graphs
sns.set(style="white", color_codes=True)

iris_data = pd.read_csv("data/iris.csv") 

# Peeking the dataset
iris_data.head()
```

### Distribution of Classes/Species in the dataset

```python
iris_data["Species"].value_counts()
```

```python
fig = plt.figure(figsize=(20,10), dpi=300) 
alpha = alpha_scatterplot = 0.2 

# lets us plot many diffrent shaped graphs together 
ax1 = plt.subplot2grid((2,3),(0,0))
iris_data.Species.value_counts().plot(kind='bar')
plt.title("Distribution of Classes/Species")
ax1.set_ylim(-1, 75)
```

```python
# Another useful seaborn plot is the pairplot, which shows the bivariate relation
# between each pair of features
# 
# From the pairplot, we'll see that the Iris-setosa species is separataed from the other
# two across all feature combinations
sns.pairplot(iris_data.drop("Id", axis=1), hue="Species", size=3)
```

```python
# We can update these elements to show other things, such as a kde
sns.pairplot(iris_data.drop("Id", axis=1), hue="Species", size=3, diag_kind="kde")
```

```python
sns.boxplot(x="Species", y='Sepal.Length', data=iris_data)
```

```python
sns.boxplot(x="Species", y='Sepal.Width', data=iris_data)
```

```python
sns.boxplot(x="Species", y='Petal.Length', data=iris_data)
```

```python
sns.boxplot(x="Species", y='Petal.Width', data=iris_data)
```

#### A final seaborn plot useful for looking at univariate relations is the kdeplot,which creates and visualizes a kernel density estimate of the underlying feature

```python
sns.FacetGrid(iris_data, hue="Species", size=6) \
   .map(sns.kdeplot, "Sepal.Length") \
   .add_legend()
```

```python
sns.FacetGrid(iris_data, hue="Species", size=6) \
   .map(sns.kdeplot, "Sepal.Width") \
   .add_legend()
```

```python
sns.FacetGrid(iris_data, hue="Species", size=6) \
   .map(sns.kdeplot, "Petal.Width") \
   .add_legend()
```

```python
sns.FacetGrid(iris_data, hue="Species", size=6) \
   .map(sns.kdeplot, "Petal.Length") \
   .add_legend()
```

###Multivariate visualization

A final multivariate visualization technique pandas has is radviz Which puts each feature as a point on a 2D plane, and then simulates
having each sample attached to those points through a spring weighted  by the relative value for that feature

```python
from pandas.tools.plotting import radviz
radviz(iris_data.drop("Id", axis=1), "Species")
```
